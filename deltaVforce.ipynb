Lancer le programme python en choisissant une force F1 √©gale √† 0,7 N et une force F2 √©gale √† 1,0 N. La masse vaut 250 g.
    Le programme affiche : la position des points ùëÄùëñ et sur chaque point, le vecteur variation de vitesse : 

```python
from matplotlib import pyplot as plt
import numpy as np

# ==== D√©finition des variables et constantes n√©cessaires ==========================
dt=0.125 # Valeur de l'intervalle de temps constant et √©gale √† 125 ms
F1 =float(input('Valeur F de la force appliqu√©e sur la premi√®re Smart Cart(entre 0.50 et 1.0 N): F1 = ? '))
F2 =float(input('Valeur F de la force appliqu√©e sur la deuxi√®me Smart Cart(entre 0.50 et 1.0 N): F2 = ? '))
m=float(input('Valeur de la  masse des Smart Cart(en kg): m = ? '))

N =20     # Nombre maximal de points √† repr√©senter

# ==== G√©n√©ration des coordonn√©es des masses ponctuelles m1 et m2 ==================
t = np.linspace(0,N*dt,N)         # D√©finition du domaine des dates (en s)
x1 = 100*((F1/(2*m))*t**2)        # D√©finition des abscisses x1 (en cm)
y1 = 100*(0*t + 0.003)            # D√©finition des ordonn√©es y1 (en cm)
x2 = 100*((F2/(2*m))*t**2)        # D√©finition des abscisses x2 (en cm)
y2 = 100*(0*t)                    # D√©finition des ordonn√©es y2 (en cm)
# distance en cm pour all√©ger la repr√©sentation graphique

# ==== Figures repr√©sentant les trajectoires y1=f(x1 et y2=f(x2) ===================
plt.figure('Positions de  m1 et m2 et vecteurs variation de vitesse ')
plt.title('Vecteurs variation de vitesse deltaV cas F1 et deltaV cas F2'+' (m='+str(round(m,3))+' kg)')
plt.xlabel('x(en cm)')
plt.ylabel('y(en cm)')
plt.plot(x1,y1,'bo',ms=2,label='Force F1='+str(F1)+' N')
plt.plot(x2,y2,'ro',ms=2,label='Force F2='+str(F2)+' N')
plt.axis([-5,100,-0.5,0.75])

# ==== Calculs des coordonn√©es des vecteurs vitesse V1 et V2 =======================
# d√©finition de 2 listes pour les variables Vx1 et Vy1
Vx1,Vx2=[],[]
for i in range(0,len(t)-1) :
    # calcul des coordonn√©es Vx1i des vecteurs vitesse sur l'axe x au point n¬∞i
    Vx1i=(x1[i+1]-x1[i])/(t[i+1]-t[i])
    Vx1.append(Vx1i)                # ajout de la valeur Vx1i √† la liste Vx1
    Vx2i=(x2[i+1]-x2[i])/(t[i+1]-t[i])
    Vx2.append(Vx2i)                # ajout de la valeur Vx2i √† la liste Vx2

# ==== Calcul des vecteur variation de vitesse deltaV1 et deltaV2=================
deltaVx1=[0]
deltaVx2= [0]
for i in range(1,len(t)-1) :
    deltaVx1i= (Vx1[i]-Vx1[i-1])
    deltaVx1.append(deltaVx1i)
    deltaVx2i= (Vx2[i]-Vx2[i-1])
    deltaVx2.append(deltaVx2i)
print("variation du vecteur vitesse pour la Smart cart soumise √† la force F1=", F1,"N, en cm/s")
print (deltaVx1)
print("variation du vecteur vitesse pour la Smart cart soumise √† la force F2=", F2,"N, en cm/s")
print(deltaVx2)
echelle=0.1
# ==== Trac√© des vecteurs variation de vitesse deltaV1 et deltaV2 ==================
for i in range(1,len(t)-1) :
    # Repr√©sente, au point d'indice i de coordonn√©es (x1[i],y1[i]), une fl√®che
    # de longueur dt*(Vx1[i]-Vx1[i-1]) sur l'axe x
    plt.arrow(x1[i],y1[i],echelle*(Vx1[i]-Vx1[i-1]),0,width=0.001,
    length_includes_head="true",head_length=1.2,head_width=0.02,color='b')
    plt.arrow(x2[i],y2[i],echelle*(Vx2[i]-Vx2[i-1]),0,width=0.001,
    length_includes_head="true",head_length=1.2,head_width=0.02,color='m')

#-affichage de l'√©chelle de repr√©sentation des vecteurs vitesse - A NE PAS MODIFIER
plt.text(25,-0.30,'Echelle du vecteur variation de vitesse :', color='green')
plt.text(27,-0.35,0.50, color='green') # affichage echelle 0,50 m/s
plt.text(30,-0.35,' m.s$^{-1}$', color='green')
# fl√®che de longueur correspondant √† 0,20 m/s
plt.arrow(50,-0.35, echelle*50, 0,width=0.005,
              length_includes_head="true", head_length=1.2,
              head_width=0.02, color='g')

plt.legend()
plt.show()
```
